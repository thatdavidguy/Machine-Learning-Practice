```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
#https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
iris = datasets.load_iris()
type(iris)
#initializing libraries and dataset, plt style chosen by me :3
import seaborn as sns
```

```python
print(iris.keys())
print(f'Data is type {type(iris.data)}')
print(f'Target is type {type(iris.target)}')
#getting the types of the dataset
```

```python
print(iris.data.shape)
#checking the shape/ ammount of features and rows/entries 
#Isn't 150 Low?
```

```python
print(iris.target_names)
#encoded as 0,1,2 (it is classification)
```

```python
X = iris.data
y = iris.target
df = pd.DataFrame(X, columns = iris.feature_names)
print(df.head())
```

```python
df.describe()
#Does not tell us much (in the objective of categorizing the different flowers)

#in terms of unsupervised learning (clustering)
#i'd assume a higher variation in the data would help differentiate the data
```

```python
_ = pd.plotting.scatter_matrix(df, c = y, figsize = [8,8],
                               s = 50, marker = '1')
plt.figure()

plt.imshow([np.unique(iris.target)])
tick = np.unique(iris.target)
_ = plt.xticks(ticks=tick,labels=iris.target_names)

# plot x against y, figure size = 8 by 8
# s is the size of the data points
# I customized my own marker :3
# Markers -> https://matplotlib.org/stable/api/markers_api.html

#To improve:
#Spacing between y intervals isnt too nice
#Legend isnt working too well :/

#EDA:
#Setosa (blue) has a smaller Sepal length, 
#versicolor(green) and virginica(red) are mixed
#For Sepal width blue is group by itself but green and red still mixed
#Petal length could be a very important feature as the groups
#are the most distinct
#Same for Petal Width
```
